```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('train.csv',sep=',')
df.head()
```

```python
df.describe()
```

```python
# mean age is 29.69. Filling the NaNs in Age with the average age of 30.
df = df.fillna(30)
```

```python
# Drop string columns that does not help with the model.
df = df.drop(df.columns[[0, 3, 8, 9, 10, 11]], axis=1)
# Use 0 to represent male and 1 to represent female.
df = df.replace({'Sex' : { 'male' : 0, 'female' : 1}})
df.head()
```

```python
# Split train and test set
X, y = df.iloc[:,1:], df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state = 42)
```

```python
ran_forest = RandomForestClassifier(n_estimators = 1, random_state = 5)
ran_forest.fit(X_train, y_train)
```

```python
# Get model accuracy
print('Accuracy on training set: {:3f}'.format(ran_forest.score(X_train, y_train)))
print('Accuracy on training set: {:3f}'.format(ran_forest.score(X_test, y_test)))
```

```python
for n in range(1,11):
    ran_forest = RandomForestClassifier(n_estimators = n, random_state = 5, max_features = 3)
    ran_forest.fit(X_train, y_train)
    print('Accuracy on training set with {} estimators: {:3f}'.format(n, ran_forest.score(X_train, y_train)))
    print('Accuracy on test set with {} estimators: {:3f}'.format(n, ran_forest.score(X_test, y_test)))
```
